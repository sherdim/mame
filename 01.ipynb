## Моделирование в биологии

*Моделирование* -  Исследование объектов познания на их моделях; построение и изучение моделей реально существующих предметов и явлений и конструируемых объектов для определения, либо улучшения их характеристик, рационализации способов их построения, управления ими и прогнозирования.

*Модель* -  Мысленный или условный образ, аналог к.-л. объекта, процесса или явления, воспроизводящий в символической форме их основные типические черты.

[Энциклопедия социологии, 2009](http://dic.academic.ru/dic.nsf/socio/2202/%D0%9C%D0%9E%D0%94%D0%95%D0%9B%D0%98%D0%A0%D0%9E%D0%92%D0%90%D0%9D%D0%98%D0%95)

# Математический язык

- насаждается как универсальный (число пи - $\pi$)
- лишен четкой грамматики, поэтому допускает двусмысленности (“=” для присвоения, идентичности, равенства)
- сложился безсистемно, поэтому изобилует дублированием и несоответствиями похожих применений

Произведение: $$ ab, a \cdot b , a * b, a \times b, a \otimes b, a \rtimes b, a \circ b, a \wedge b,  \prod_{i \,\in\, \Omega} a_i $$


Полезное: [угадать символ по рисунку](http://detexify.kirelabs.org/classify.html),    [найти в таблице](https://ru.wikipedia.org/wiki/%D0%A2%D0%B0%D0%B1%D0%BB%D0%B8%D1%86%D0%B0_%D0%BC%D0%B0%D1%82%D0%B5%D0%BC%D0%B0%D1%82%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%B8%D1%85_%D1%81%D0%B8%D0%BC%D0%B2%D0%BE%D0%BB%D0%BE%D0%B2)

число пи - во всех языках $\pi$, иначе только "отношения длины окружности к диаметру"

$times$  - декартово произведение множеств

$\otimes$  - Тензорное произведение для множеств, для векторов - Произведение Кронекера

$\rtimes$  - Полупрямое произведение

$\circ$  - Композиция функций

$\wedge$ - логическое умножение

Буква П - произведение, если начинается с единицы, то есть специальное обозначение факториал в виде восклицательного знака.

главный недостаток - произвольное изобретение разными математиками символов, когда возникает потребность - со временем один из вариантов получает распространение

## Функция = соответствие между двумя множествами с заданным направлением

### 2. Математическое понятие

*Отображение* одного множества на другое,  Преобразования элементов одного множества в элементы другого, *Оператор*

Непрерывная функция - абстракция, когда любое изменение аргумента приводит к изменению значений.

*Функцией $f$* называется множество упорядоченных пар $(x,y)\in X\times Y$, таких, что пары существуют для всех элементов множества $X$, и, если первые элементы пар совпадают, то совпадают и вторые их элементы.

При этом:
- Множество $X$ называется областью задания функции, домен;
- множество $Y$ называется областью значений функции, ранг.

## Алгоритм использования математических методов для биолога

1. Если ПО строит нужный рисунок - используй его.
2. Если ПО дает данные для нужного рисунка - используй его и настрой параметры рисунка.
