# Проектное задание

проектное задание представляет собой маленькую исследовательскую работу на тему «Исследование динамических свойств внимания по параметрам выполнения сенсомоторных реакций».

В ходе работы над проектом студенты вырабатывают навыки:

  - работы с литературой для выбора сценария исследования;
  - планирования эксперимента;
  - создания базы данных путем прохождения психофизиологического тестирования;
  - анализа данных;
  - представления данных;
  - формирования отчета о научно-исследовательской работе;
  - выступления с докладом на защите проекта.


## Креатифф

Проектное задание трудоемкое, требует творческого подхода и внимательности. В ходе выполнения Вам необходимо много раз сделать несложный, но самостоятельный выбор:

-    какие условия выбрать для сценария тестирования;
-    как назвать переменные (lyuboe_sochetanie_latinskih_bukv);
-    какой из алгоритмов выбрать на каждом из этапов обработки данных;
-    какой цвет выбрать для рисунка;
-    влияние каких факторов анализировать (при оценке влияния одного фактора, градации всех остальных должны совпадать);
-    методично и плодотворно подойти к работе или оставить все на последний вечер;
-    самому вникать в подробности программирования или прийти на консультацию...

## План работы

1. Выбрать сценарий исследования.
1.      Получить данные с помощью теста "Баланс внимания" (или Ускорение или другие).

2.      Обработать данные с помощью освоенных методик.

3.      Сформировать отчет с иллюстрациями.

### Выбор сценарий исследования

- что вам интересно?
- где в вашу жизнь можно внедрить тестирование?
- заявить о своем сценарии на форуме

### Сообщение на форуме

- перечислите условия тестирования, которые будут сравниваться между собой. 
- кратко изложите рабочую гипотезу, как по Вашему мнению эти условия могут повлиять на механизмы внимания.

Не откладывайте выбор методики тестирования на потом. Ранее продумывание схемы эксперимента дает следующие преимущества:
-	преподаватель может сделать замечание и подсказать более предпочтительный вариант тестирования до того, как Вы приступите к работе;
-	публикацией выбранной методики Вы закрепляете ее за собой. Повторять тестирование другим студентам по тому же самому сценарию категорически не рекомендуется.


## Набор экспериментальных данных

[Приложение "Баланс внимания"](http://balatte.stireac.com/) или [Приложение "Ускорение"](http://uskor.stireac.com/)
используется для самостоятельного выполнения проектного задания

Тест желательно проходить в браузере Google Chrome на настольном компьютере или ноутбуке, т.е. где есть мышка и достаточная производительность.

Студент проходит тестирование в нескольких условиях, о чем делает соответствующие пометки. 


### Обработка данных
- длинный структурированный блокнот
  - идеи
  - рисунки
  - выводы

### Рисунки из приложения не годятся??
Поскольку надо будет сравнить результаты нескольких тестирований, то рисунков, которые предлагаются разработчиками приложения, недостаточно. 

На итоговом рисунке должны быть наглядно представлены данные, полученные в разных условиях. 

### Оформление отчета в форме статьи

- введение (идея, цель), 
- методика (условия тестирования, под каким логином проходили тестирование, использованные методы анализа данных), 
- результаты (рисунки, описание), 
- выводы.

### Визуализация
Поскольку вам надо будет сравнить результаты нескольких тестирований, то рисунков, которые предлагаются разработчиками приложения, недостаточно. На итоговом рисунке должны быть наглядно представлены данные, полученные в разных условиях.
