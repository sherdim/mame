# Квалиметрические модели

**Качество** - совокупность характеристик объекта, определяющих его способности удовлетворять определенные потребности в соответствии с назначением.

Оценка качества продукта или операции основана на сопоставлении наличных результатов с неким эталоном.
Эталон описывается набором показателей, по которым его можно сравнить с пробными образцами.

Сравнение производят по критериям.
Критерий - это правило получения качественной оценки (годен/негоден) из измеряемого показателя.
Оценки также могут быть экспертными. В этом случае вместо измерений берут в рассчет суждения нескольких специалистов и используют наиболее часто встречающееся из них.

Простейшим критерием для измеряемого показателя может быть сравнение с пороговой величиной.
Например, рост в холке взрослой собачки меньше 18 см, значит особь не соответствует стандартам породы Пражский крысарик.

Результат сравнения показателя $p$ качества пробного образца с соответствующим показателем базового образца (эталона) выражают в виде некоторой безразмерной величины, которую называют оценкой показателя $Q_p$.



В этой практической работе Вы освоите:
- комбинирование количественных и экспертных оценок;
- определение критериев качества;
- синтез комплексных показателей.
