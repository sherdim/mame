# Как работать с блокнотами упражнений?

В ресурсах к пособию можно найти блокноты с заготовками для упражнений в решении различных прикладных задач в программной среде Jupyter Notebook.

Блокноты упражнений начинаются с префикса `Task_`.
Заготовки для контрольных работ имеют специальные префиксы.

Содержание блокнотов с образцами кода может быть самым разным.

Начинаются блокноты с заголовка - это ячейка типа Markdown, текст которой предваряет символ `#` (диеза) (`Shift-3' в латинской раскладке).

```py
# Заголовок блокнота
```

Комментарии написаны обычным текстом.

Инструкции или вопросы для самоконтроля обычно выделены в форме цитат (с отступом и серой вертикальной полоской слева).

В качестве примера цитаты приведем заголовок статьи, в которой было дано описание блокнотов.

> Jupyter Notebooks – a publishing format for reproducible computational workflows
*http://ebooks.iospress.nl/publication/42900*

*Перевод*: Юпытер Блокноты - формат публикации для воспроизводимых вычислительных рабочих процессов

Рабочий процесс (workflow) - это конкретное воплощение алгоритма (-ов).
Готовый отлаженный блокнот, представляющий собой набор ячеек с кодом, выполняемых друг за другом, и составляет основу для запуска рабочего процесса обработки данных.

Если блокнот длинный, то группу ячеек, объединенных одной идеей, предваряют заголовком второго уровня, для чего используются два символа `#`:

```py
## Подзаголовок 1
```

Можно также создать сложную иерархию из заголовков 3-го, 4-го и более уровней.

Адреса веб-ресурсов в тексте  (начинается с `http://` или `https://`) автоматически преобразуются в кликабельную ссылку. 
Вы можете видеть такую в предыдущей цитате.

В следующей цитате ссылка сделана из подписи и адреса по следующей схеме:

```py
[Видимый текст ссылки в квадратных скобках](http://адрес-веб-ресурса/в/круглых/скобках)
```
Например, 
```
[Jupyter Notebook](http://jupyter.org/)

```

> The [Jupyter Notebook](http://jupyter.org/) is a web application that allows you to create and share documents that contain live code, equations, visualizations and explanatory text. 

Перевод: [Юпытер Блокнот](http://jupyter.org/) - это веб-приложение, которое позволяет создавать и совместно использовать документы, которые содержат живой код, уравнения, визуализации и пояснительный текст.

Похожий синтаксис используется для быстрой вставки картинок. Следующая картинка вставлена с помощью кода:

```md
![](https://jupyter.org/assets/nav_logo.svg)
```

В данном случае квадратные скобки пустые, и перед ними стоит восклицательный знак.
Такая конструкция преобразуется в картинку, которая подгружается из указанного источника: это может быть адрес в Интернете (как в этом случае) или ссылка на файл на локальном диске компьютера.

![](https://jupyter.org/assets/nav_logo.svg)

Когда вам нужно поделиться результатами своей работы, электронный блокнот с результатами анализа может быть сохранен в виде HTML-файла, который можно послать получателю напрямую, либо разместить на веб-сайте как статическую веб-страницу.
