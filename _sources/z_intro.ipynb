На этом практическом занятии мы освоили работу со скриптбуками.

Jupyter Notebook - самый популярный движок скриптбуков.

Python - самый популярный язык при работе в скриптбуках.

Если Вы попросите пример алгоритма какого-нибудь генетического анализа у какого-нибудь коллеги в мире, то скорее всего Вам пришлют блокнот в формате Jupyter Notebook с расширением ipynb. И открыв этот блокнот, Вы сможете подставить туда свои данные и повторить команды с использованием указанных в ячейках функций.

> Почему такой подход популярен среди ученых?

В обычном программировании программа должна быть заранее спроектирована, затем написана и отлажена.
В скриптбуках можно начинать работу, не зная заранее, чем закончится анализ данных. Можно загрузить данные и изучать, пробовать, изобретать новые способы анализа.
При этом Python - это полноценный язык программирования.
Можно работать с любыми данными достаточно больших объемов, со звуком, с картинками и т.п.
По своим возможностям эта среда намного превосходит электронные таблицы (Excel).
