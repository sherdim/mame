# N-мерные массивы данных

Классические измерения 0-мерны. Каждое числовое значение (datum, данное) снабжается набором метаданных.
Метаданные используются для группировки и сравнения наборов данных (data, данные).

Когда измерения выстроены в определенном порядке друг относительно друга, они формируют естественный ряд. Элементы ряда могут быть помечены "1-е сутки", "2-е сутки" и т.д.
или "вибрисса 1", "вибрисса 2" и т.д.
В любом случае положение измеряемого объекта детерминировано относительно соседних.

Временные ряды - естественные 1-мерные массивы.
Поэтому повторные измерения всегда добавляют одно измерение к базовой размерности.

2-мерный массив данных получается при фиксации значений в плоскости.
Если мы делаем одновременные измерения в нескольких точках плоскости, то получаем ествественный двумерный массив.

Огромное количество приборов используют двумерный сенсор для фиксации значений. Особенно для фиксации плотности потока фотонов в разного рода фотоматрицах (фотокамера).



В этой практической работе Вы освоите:
- работу с двумерными массивами данных;
- способы преобразований матричных данных;
- анализ цветных фотоизображений.
