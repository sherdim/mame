# Линейные модели

Если принять допущение, что чем больше действие фактора, тем больше эффект, то такие завимости имеют вид линии, угол которой соответсвует пропорции между фактором и эффектом от его действия. Модели, которые принимают это допущение, называют линейными.

Семейство линейных моделей очень обширно: от простейших до иерархических сложных. Эти модели получили широкое распространение из-за простоты расчетов и легкости интерпретации.

В этой практической работе Вы освоите:
- сравнение групповых выборок;
- сравнение зависимостей;
- основы вариационного анализа.
