# Нелинейные модели

В этой практической работе Вы освоите:
- подходы к анализу временных рядов и сигналов;
- способы разложения сложных процессов на компоненты;
- подходы к прогнозированию.

Модели процессов, которые плохо описываются линейным графиком (а в биологии это почти все процессы), требуют иной формулировки, нежели сумма действия нескольких факторов.

Подходы для описания таких моделей заимствованы из радиофизики, химии и других направлений. Наиболее сложные модели работы молекулярных систем,  взаимодействия популяций и сложного поведения животных разработаны в рамках смежных дисциплин, где для решения задач биологии привлекается опыт математиков и физиков. 

Некоторые биологические процессы удалось объяснить как сочетание нескольких процессов, изменяющихся по экспоненте: потенциал действия в нейроне, пятна на шкуре леопарда и т.п.

Другие популярные модели биофизиков изящно описывают частные наблюдаемые явления в живой природе, но ненадежны для предсказаний в реальных изменяющихся условиях. Например, колебания в модели хищник-жертва - фантом. Любое незначительное влияние остановит этот маятник: эпидемия или нарушение ареала обитания (например, строительство дороги, через которую зайцы будут перебегать, а рыси нет).
Т.е. колебания есть в наблюдаемом статистическом сигнале, но их нет в источнике сигнала.

Часто биологические процессы настолько сложные, что их нельзя редуцировать к сумме физическо-химических процессов вроде затухающих колебаний.

Поведение примитивных сообществ животных оказывается способным решать многие задачи, с которыми плохо справляются даже суперкомпьютеры, если решать проблемы в лоб - перебором.
Тогда живые процессы становятся источником вдохновения для специалистов в области информационных технологий.

Пример. Эволюционные вычисления с использованием продвинутого алгоритма кормежки бактерий  используется для анализа медицинских изображений. Оптимизация получения 3-хмерных медицинских изображений - томограмм - настолько сложна, что используются математические модели, "подсмотренные" у эволюционного отбора и коллективного поведения простейших {cite}`bermejo2015`.


> Если биолог оценивает жизнь как чудо, то это настоящий биолог. Если биолог изучает устройство биологических объектов как механических систем, то это физик или химик.
