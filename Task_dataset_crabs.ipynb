# Загрузка данных о крабах

Один из наборов данных, идущих с пакетом MASS на R - [описание](https://vincentarelbundock.github.io/Rdatasets/doc/MASS/crabs.html)

Содержит морфологические измерения крабов двух цветовых форм Leptograpsus variegatus, собранных в Австралии.

![](crabiki.jpg)

Колонки:
- sp - разновидность: "B" - blue, голубой, "O" - orange, оранжевый.
- sex - пол: F - female; M - male.
- index - номер в 1 из 4 групп 
- FL - размер лобной доли (мм)
- RW - ширина зада (мм)
- CL - длина карапакса (мм)
- CW - ширина карапакса (мм)
- BD - глубина тела (мм)


Первоисточник

Campbell, N.A. and Mahon, R.J. (1974) A multivariate study of variation in two species of rock crab of genus Leptograpsus. Australian Journal of Zoology 22, 417–425. 



```python
%matplotlib notebook
import pandas as pd
import seaborn as sns
D = pd.read_csv('crabs.csv', index_col=0)
D.info()
```

Посмотрим распределение пропорций карапакса.

```python
sns.lmplot('CL', 'CW', data=D, hue='sex');
```
