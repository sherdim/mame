## Математический минимум для биолога
- операции с матрицами (*линейная алгебра*) - для работы с массивами данных

## Математическая биология

Математическая биология - раздел математики.
**Вычислительная биология** — это междисциплинарный подход, использующий достижения информатики (и вычислительной техники), прикладной математики и статистики для решения проблем, поставленных биологией. Главными областями в биологии, которые применяют такие методы, являются:
- Биоинформатика 
- Молекулярное моделирование (НИИФОХ)
- структурная геномика 

Математическая биология полноценно изучает биохимические, биофизические процессы.

## Алгоритм использования математических методов для биолога

1. Если ПО строит нужный рисунок - используй его.
2. Если ПО дает данные для нужного рисунка - используй его и настрой параметры рисунка.
1. Если известен метод, найди реализующее его ПО, настрой загрузку своих данных и выдачу данных для нужного рисунка.
1. Если есть описание алгоритма, найди реализующий его модуль, доработай его под свои нужды.
1. Если алгоритм никто не реализовал, напиши модуль сам. Проверка работоспособности модуля - отдельная работа.
1. Если алгоритм не известен, придумай его, создай новое научное направление.

Мораль - математики уже все придумали. Надо грамотно этим пользоваться.

Углубляться в мат. методы необходимо по мере усложнения задач.

## Концепция

разрешение противоречия заключается в повышении уровня владения информационными технологиями: от манипулирования данными на низком уровне, которым по праву должны заниматься специалисты-компьютерологи, биологи переходят к пользованию высокоуровневыми пакетами математической обработки

![](softup.png)

Базовый уровень информационных технологий - калькулятор.

таблички - программирование заключается в расположении ячеек и вставке функций

...

Самый высокий уровень - готовое приложение, которое идет к прибору.
Обычно можно докупить модули. Но весь нужный функционал уже готов - ничего программировать не надо. Удобно для врачей и "пользователей".

В науке нужна новизна, а значит новые алгоритмы обработки, которых готовых нет. Поэтому мы отступаем шаг назад - скриптбук.

## Противоречие научной новизны и пользователя ПО

- Биология направлена на новые достижения. 
 - Новые задачи порождают развитие новых методов. 
   - Сложные задачи порождают новые направления математики
     - Математики решают новые абтрактные задачи

| |
-|-
Для решения новой задачи надо новые алгоритмы, т.е. делать "под себя". | vs | Для повседневного решения - нужна автоматизация, т.е. готовое приложение.


Новый алгоритм = конструирование из готовых блоков.

Необходимо:
- знать какие бывают блоки,
- уметь их стыковать,
- знать и уметь проверить работоспособность конструкции.
