## Моделирование в биологии

*Моделирование* -  Исследование объектов познания на их моделях; построение и изучение моделей реально существующих предметов и явлений и конструируемых объектов для определения, либо улучшения их характеристик, рационализации способов их построения, управления ими и прогнозирования.

*Модель* -  Мысленный или условный образ, аналог к.-л. объекта, процесса или явления, воспроизводящий в символической форме их основные типические черты.

[Энциклопедия социологии, 2009](http://dic.academic.ru/dic.nsf/socio/2202/%D0%9C%D0%9E%D0%94%D0%95%D0%9B%D0%98%D0%A0%D0%9E%D0%92%D0%90%D0%9D%D0%98%D0%95)

#### Пример конкретизации модели. Действие наперстянки

Британские ученые задокументировали [историю](http://www.historyofscience.com/articles/jmnorman-william-withering.php) времен Михаила Ломоносова, когда некая знахарка [Матушка Хаттон](https://en.wikipedia.org/wiki/Mother_Hutton) рассказала секрет снадобья сэру Уильяму Визирингу, ботанику, который впервые провел клинические испытания на группе пациентов, о чем опубликовал. Кстати, дедушка Чарльза Дарвина, Эразм Дарвин обвинял его в переманивании пациентов, поскольку тоже опубликовал работу о успешном лечении пациентов препаратами наперстянки в том же 1785 году. [подробнее](http://www.sciencedirect.com/science/article/pii/S0735109785804575)

[Наперстянка](https://ru.wikipedia.org/wiki/%D0%9D%D0%B0%D0%BF%D0%B5%D1%80%D1%81%D1%82%D1%8F%D0%BD%D0%BA%D0%B0) (лат. Digitális purpúrea) ![](http://fitoapteka.org/images/foto/629/18.jpg)

Наперстянка - смертельно ядовитое растение. Отравление сопровождается аритмией, тошнотой, рвотой, поносом, посинением губ и удушьем.

Наперстянка входила в качестве ингридиента в грудной сбор из более 20 трав. Визиринг выявил ключевую роль именно наперстянки, проверив эффект от разных частей растения, собранных в разное время года.

Была создана модель: если пациенту с отеком легких дать порошок из измельченных листьев дигиталиса, то ему полегчает.

Модель можно дополнить, что там содержится некое вещество, которое заставляет сердце работать нормально. При этом влага из груди "откачивается". Например, лучше выделяется почками, т.е. диуретическое действие.

Уабаин и его аналоги называют *кардиотонические стероиды* или *сердечные гликозиды*, а в тканях наперстянки накапливаются один из них - дигитоксин (от родового названия дигиталис). 

<img style=float:right src=uabain_model.png>
Современная модель следующая: уабаин связывается с Na,K-АТФазой, являясь ее специфическим блокатором. Каскад воздействий: Сердечный гликозид, блокирует транспортную Na+/K+-АТФ-азу, в результате чего возрастает содержание Na+ в кардиомиоците, что приводит к открытию Ca2+-каналов и вхождению Ca2+ в кардиомиоциты. Повышение концентрации Ca2+ приводит к ингибированию тропонинового комплекса, оказывающего угнетающее влияние на взаимодействие актина и миозина. В результате увеличения сократительной способности миокарда увеличивается ударный объем кровотока и минутный объем кровотока (МОК). Снижаются конечный систолический объем и конечный диастолический объем сердца, что наряду с повышением тонуса миокарда приводит к сокращению его размеров и т.о. к снижению потребности миокарда в кислороде. Благодаря раздражению рецепторов в сердце (кардиокардиальный рефлекс), усилению систолического выброса и раздражению барорецепторов дуги аорты и сонного клубочка, повышается тонус блуждающего нерва. Это приводит к снижению активности синоатриального узла (отрицательное батмотропное действие) и атриовентрикулярного узла (отрицательное батмо- и дромотропное действие). На фоне отрицательного хронотропного эффекта и положительного инотропного эффекта систола становится более короткой и энергетически экономичной. Сердечный ритм замедляется.

## Язык и письмо

Стыкующий элемент - алфавит (dictionary) - определяет соответствие между символом и понятием (обозначением и денотатом)

Знание письменного языка - умение из прочитанных символов извлечь заложенное в них понятие. Письменность - пока еще основной способ передачи знаний.

У языка может быть несколько письменностей, и одна письменность для многих языков.

**Письмо**:
```
латиница:  Statistics  Statistique  Statystyka  Estatistika
эллиница (-900):   Στατιστική
евраица (-850):      סטטיסטיקה
эфиопица:  እስታቲስቲክስ     [' s ta ti s ti k s]
деванагари: सांख्यिकी
тамилица:   புள்ளியியல்  
арабица:    إحصاء     شماریات
мхедрули (405):   სტატისტიკა
кириллица (863):  Статистика  Статыстыка
```

### Математическая модель - модель, при описании которой использован математический язык

Математическую модель можно записать на языке формул или языке программирования.

Итак, мы уже определелись, что хотим описать модель на математическом языке. Этого требуют традиция и желание быть точным или корректно неточным.

#### Модель - лишь подобие реальности. 
Математическая модель сопровождается:
- оценкой сходства - например, точность - процент случаев, модель совпадает с реальностью.
- границами применимости - описанием условий, в которых модель ведет себя адекватно.
- примером использования - важно для сравнения с другими моделями

Содержательно, математическая модель - это формула или программа (последовательность команд).

Если *Модель* описывает взаимосвязи между элементами системы (живой),то *Математическая модель* описывает отношения между измерямыми характеристиками этих элементов.

Модель описывает взаимосвязи между элементами системы (живой). Математическая модель описывает отношения между измерямыми характеристиками этих элементов.

## Функция = соответствие между двумя множествами с заданным направлением

### 1. Общефилософское понятие

Функция клавиатуры - между нажатиями на кнопки и вводимыми в редакторе символами.

Функциональная клавиша - можно присвоить разные функции в разных приложениях.

## Функция = соответствие между двумя множествами с заданным направлением

### 2. Математическое понятие

*Отображение* одного множества на другое,  Преобразования элементов одного множества в элементы другого, *Оператор*

Непрерывная функция - абстракция, когда любое изменение аргумента приводит к изменению значений.

*Функцией $f$* называется множество упорядоченных пар $(x,y)\in X\times Y$, таких, что пары существуют для всех элементов множества $X$, и, если первые элементы пар совпадают, то совпадают и вторые их элементы.

При этом:
- Множество $X$ называется областью задания функции, домен;
- множество $Y$ называется областью значений функции, ранг.

Математическая модель - математическое описание фукнциональной связи. 
Позволяет предсказывать поведение системы при изменении факторов.
