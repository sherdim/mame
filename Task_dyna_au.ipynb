```python
from scipy.io import wavfile
```

```python
u='eegfilt.wav'
freq,y = wavfile.read(u)
y.shape
```

```python
fig, (ax1,ax2) = subplots(2,1, figsize=(10,6))
x = arange(len(y))/freq
ax1.plot(x,y)
ax2.specgram(y,Fs=freq);
```
