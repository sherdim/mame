# Нахождение максимума функции

На опушке поля живет колония сусликов. У колонии много выходов, при этом колония занимает на земле круг диаметром 5 м.
В 3 м от крайнего выхода растет высокое сухое дерево высотой более 10 м.
Сокол, сидя на дереве, высматривает себе добычу.

Согласно модели оптимальная высота ветки для посадки сокола соответствует наиболее широкому углу обзора, чтобы в поле зрения попадала вся колония с максимальным разрешением. Разрешение изображения тем выше, чем ближе к объекту, но если сокол будет сидеть совсем низко - будет видеть колонию сбоку как узкий овал, т.е. угол сузится. Необходимо найти оптимальную высоту ветки для наблюдения за колонией сидящему на дереве хищнику.

Предположим, что уровень глаз птицы находится на 10 см выше ветки.

$$\theta = \alpha - \beta = atg\left(\frac{8}{x}\right) - atg\left(\frac{3}{x}\right)$$
$$x=?\ ,\theta \rightarrow \max$$

```python
%pylab inline
```

```python
d=5.0
s=3.0
x=arange(.1,10,.01)
a=arctan((d+s)/x)
b=arctan(s/x)
th=a-b
plot(x,a,'b', x,b,'r', x,th,'g');
xlabel('Высота, м'); ylabel('Угол, рад'); legend(['$alpha$', '$beta$','$theta$']);
```

Находим позицию массива, где угол $\theta$ максимальный. И определяем, чему равен $x$ в этой позиции

```python
x[argmax(th)]
```

Значение в радианах преобразовываем в градусы - делим на $\pi$ и умножаем на 180 градусов.

```python
rad2deg(max(th))
```
