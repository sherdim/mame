# Самостоятельная работа "Сравнение регрессионных моделей"

В этой работе Вы должны продемонстрировать умение рассчитывать линейные регрессионные модели и сравнивать их между собой.

Образцы кода, который можно использовать для формирования отчета, основаны на данных из приложения "Баланс внимания". Желающие могут использовать свои данные или данные из проектов, в которых они участвовали. 


### Требования к данным
1. Две или более подвыборки с количеством значений не менее 15 в каждой.

Все этапы расчетов должны сопровождаться наглядными иллюстрациями и пояснениями.

## Про коэффициент детерминации

Для рассчета нужно найти дисперсию данных и дисперсию остатков. 

$$R^2 = 1 - \frac{\sigma_r^2}{\sigma^2}$$

Поскольку обе сигмы делятся на одно и то же n, то можно поделить сумму квадратов разницы с модельными значениями на сумму квадратов разницы со средним.

$$R^2 = 1 - \frac{sum((y-\hat{y})^2)}{sum((y-\bar{y})^2)}  = corrcoef(x, y)^2$$
