# Загрузка данных о весе кошек

Один из наборов данных, идущих с пакетом R - [описание](https://vincentarelbundock.github.io/Rdatasets/doc/MASS/cats.html)

Данные по влиянию дозы препарата наперстянки на сердечную деятельность.

- Sex - пол: "F" или "M".
- Bwt - масса тела в кг.
- Hwt - масса сердца в г.


Первоисточник

R. A. Fisher (1947) The analysis of covariance method for the relation between a part and the whole, Biometrics 3, 65–68.


```python
%matplotlib notebook
import pandas as pd
import seaborn as sns
D = pd.read_csv('cats.csv', index_col=0)
D
```

Посмотрим распределение значений.

```python
sns.lmplot('Bwt', 'Hwt', data=D, hue='Sex');
```
